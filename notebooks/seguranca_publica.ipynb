# Projeto de Banco de Dados II

Nesse trabalho, faremos uma análise da segurança pública com enfoque na
exposição dos crimes cometidos no estado de São Paulo.

## Justificativa

Para o eleitorado brasileiro, um dos principais itens a ser levado em
consideração na hora de decidir seu próximo representante é a segurança pública,
seguido de itens como saúde e economia. Segundo uma pesquisa do [DataFolha](https://www.cnnbrasil.com.br/nacional/sudeste/sp/datafolha-percepcao-de-violencia-e-temor-de-assaltos-e-maior-em-sao-paulo/), a
sensação de insegurança e aumento da violência entre os moradores de São Paulo
é maior que toda a média brasileira, mesmo que os dados oficiais da Secretaria de [Segurança Pública](https://www.ssp.sp.gov.br/estatistica/dados-mensais) (SSP) mostrem que crimes violentos, como homicídio, tiveram
uma diminuição nos casos nos últimos anos.

Elucidado dos problemas supracitados, nosso objetivo é mostrar, por meio dos
dados, como se encontra a segurança na cidade e, possivelmente, no estado de
São Paulo, expondo de forma fácil e visual quais os crimes cometidos, onde eles
estão ocorrendo e como se encontra o policiamento em determinada região, além,
de se possível, realizar uma previsão da taxa de possíveis delitos em determinada área

## Objetivos

Nosso objetivo principal é a **criação de uma visualização fácil e rápida dos
dados tangentes à criminalidade em São Paulo capital** e, se possível, no estado.
Nosso foco é
1. Entender as variáveis que influenciam no cumprimento de uma
infração da lei
2. Melhorar a visualização dessas informações para o público geral.

Como objetivo secundário, foca-se em possíveis soluções para melhorar a
taxa da criminalidade - ou de determinado crime - em alguma região. Elucida-se também na criação de diferentes mapas com filtros para tipo de crime, IDH;
ranqueamento de crimes mais recorrentes; correlação de crimes com região;
sensação de segurança com patrulhamento policial e outras métricas.

## Fonte de dados

### 1° - Segurança Pública de São Paulo (SSP)

A primeira fonte será oficial da [SSP](https://www.ssp.sp.gov.br/). Iremos recuperar, por meio do portal de transparência, dados gerais sobre os crimes cometidos, quantidade, locais, diferença temporal

### 2° - Observa Sampa

A segunda fonte a ser analisada é o [ObservaSampa](https://dados-abertos-observasampa.prefeitura.sp.gov.br/), que é o Observatório de Indicadores da Cidade de São Paulo e tem como objetivo apresentar dados, indicadores e evidências sobre a cidade de São Paulo e suas políticas públicas, por meio de estudos, publicações, dados abertos e outros.

### 3° - Criminalidade bairro a bairro

A terceira se trata de um mapa interativo da criminalidade bairro a bairro feito pelo [Estadão](https://infograficos.estadao.com.br/cidades/criminalidade-bairro-a-bairro/). Apesar da fonte deles serem o SSP - já usado anteriormente -, nossa tarefa será fazer um web scrapping para retirar as informações do site que possam ser interessante na nossa análise.
